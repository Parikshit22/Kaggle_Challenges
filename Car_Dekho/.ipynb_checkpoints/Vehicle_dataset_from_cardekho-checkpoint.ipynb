```python
import pandas as pd
```

```python
df = pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Year"] = df["Year"].apply(lambda x : 2020-x)
```

```python
df.drop(["Car_Name"],axis = 1,inplace = True)
```

```python
df = pd.get_dummies(df,drop_first = True)
```

```python
df.head()
```

```python
df.corr()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df)
```

```python
import matplotlib.pyplot as plt
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
Y = df.loc[:,"Selling_Price"]
X = df.drop("Selling_Price",axis = 1)
```

```python
X.columns
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,Y)
```

```python
print(model.feature_importances_)
```

```python
feature_importance = pd.Series(model.feature_importances_,index = X.columns)
# feature_importance.plot(kind = 'barh')
feature_importance.nlargest(5).plot(kind = 'barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
import numpy as np
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = regressor, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
prediction = rf_random.predict(x_test)
```

```python
prediction
```

```python
sns.distplot(y_test-prediction)
```

```python
import pickle
with open("cardekho_random_foreset_regression_model.pkl",'wb') as f:
  pickle.dump(rf_random,f) 
```
